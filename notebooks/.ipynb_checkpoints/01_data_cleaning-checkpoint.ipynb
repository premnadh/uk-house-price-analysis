```python
import pandas as pd

file_path = "../data/raw/UK-HPI-full-file-2025-05.csv"

df = pd.read_csv(file_path)

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].dtype
```

```python
df = df.sort_values('Date')
df = df.reset_index(drop=True)
```

```python
cols_to_keep = [
    'Date','RegionName','AveragePrice','SalesVolume',
    'DetachedPrice','SemiDetachedPrice',
    'TerracedPrice','FlatPrice',
    'CashPrice','MortgagePrice'
]

housing = df[cols_to_keep]
housing.head()
```

```python
housing.isna().sum()
```

```python
housing = housing.ffill()
housing = housing.bfill()
```

```python
housing.isna().sum()
```

```python
housing.to_csv("../data/processed/cleaned_housing.csv", index=False)
```
